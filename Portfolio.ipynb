```python
import yfinance as yf
import pyfolio as pf 
import warnings
import pandas as pd

warnings.filterwarnings('ignore')
btc = yf.Ticker("BTC-USD")
eth = yf.Ticker("ETH-USD")
iau = yf.Ticker("IAU")
sil = yf.Ticker("SIL")
slv = yf.Ticker("SLV")
itsa = yf.Ticker("ITSA4.SA")
egie = yf.Ticker("EGIE3.SA")
spy = yf.Ticker("SPY")
knri = yf.Ticker("KNRI11.SA")
xplg = yf.Ticker("XPLG11.SA")
xpml = yf.Ticker("XPML11.SA")
dolar = yf.Ticker("USDBRL=X")
```

```python
xplgdf
```

```python
def dfs(df):
    df2 = df.history(period="max")
    df2['returns'] = df2.Close.pct_change(1)
    return df2
btcdf = dfs(btc)
ethdf = dfs(eth)
iaudf = dfs(iau)
sildf = dfs(sil)
slvdf = dfs(slv)
itsadf = dfs(itsa)
egiedf = dfs(egie)
spydf = dfs(spy)
knridf = dfs(knri)
xplgdf = dfs(xplg)
xpmldf = dfs(xpml)
dolardf = dfs(dolar)

#sharpe = btcdf['return'].mean() /btcdf['return'].std() 
#sharpe = sharpe * 252**0.5
#sharpe
```

```python
pf.create_full_tear_sheet(dolardf['returns'])
```

```python
portf= pd.concat([btcdf.rename(columns={'Close':'BTC'}).BTC,ethdf.rename(columns={'Close':'ETH'}).ETH,iaudf.rename(columns={'Close':'IAU'}).IAU,sildf.rename(columns={'Close':'SIL'}).SIL,slvdf.rename(columns={'Close':'SLV'}).SLV,itsadf.rename(columns={'Close':'ITSA'}).ITSA,egiedf.rename(columns={'Close':'EGIE'}).EGIE,spydf.rename(columns={'Close':'SPY'}).SPY,dolardf.rename(columns={'Close':'DOLAR'}).DOLAR],axis=1)
portf = portf.dropna()
```

```python
portf['BTCBRL'] = portf.BTC * portf.DOLAR
portf['ETHBRL'] = portf.ETH * portf.DOLAR
portf['SILBRL'] = portf.SIL * portf.DOLAR
portf['SLVBRL'] = portf.SLV * portf.DOLAR
portf['IAUBRL'] = portf.IAU * portf.DOLAR

portf = portf[['BTCBRL','ETHBRL','SILBRL','SLVBRL','IAUBRL','ITSA','EGIE',]]
```

```python
rets= pd.concat([btcdf.rename(columns={'returns':'BTC'}).BTC,ethdf.rename(columns={'returns':'ETH'}).ETH,iaudf.rename(columns={'returns':'IAU'}).IAU,sildf.rename(columns={'returns':'SIL'}).SIL,slvdf.rename(columns={'returns':'SLV'}).SLV,itsadf.rename(columns={'returns':'ITSA'}).ITSA,egiedf.rename(columns={'returns':'EGIE'}).EGIE,knridf.rename(columns={'returns':'KNRI'}).KNRI,xplgdf.rename(columns={'returns':'XPLG'}).XPLG,xpmldf.rename(columns={'returns':'XPML'}).XPML,spydf.rename(columns={'returns':'SPY'}).SPY,dolardf.rename(columns={'returns':'DOLAR'}).DOLAR],axis=1)
#rets = rets.dropna()
```
